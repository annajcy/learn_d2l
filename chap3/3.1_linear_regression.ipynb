```python
import math
import torch
import numpy as np
import time

from d2l.plot import plot, set_axes
import matplotlib.pyplot as plt
```

```python
n = 10000
a = torch.ones(n)
b = torch.ones(n)
```

```python
c = torch.zeros(n)
t = time.time()
for i in range(n):
    c[i] = a[i] + b[i]
f'{time.time() - t:.5f} sec'
```

```python
def normal(x, mu, sigma):
    p = 1 / math.sqrt(2 * math.pi * sigma**2)
    return p * np.exp(-0.5 * (x - mu)**2 / sigma**2)
```

```python
x = np.arange(-7, 7, 0.1)

parameters = [(0, 1), (0, 2), (3, 1)]
fig, ax = plt.subplots()
plot(ax, 
     ([x, x, x], [normal(x, mu, sigma) for mu, sigma in parameters]),
     ('x', 'p(x)'),
     ((-7, 7), (0, 0.5)),
     ['normal(0, 1)', 'normal(0, 2)', 'normal(3, 1)'])
```
